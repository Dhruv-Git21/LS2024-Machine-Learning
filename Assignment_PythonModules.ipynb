```python
import pandas as pd 

## Create a DataFrame from a dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
#Your code here
data_df=pd.DataFrame(data)

#Display the first 2 rows of the data frame
print(data_df.head(2))

#Print the age column
print(data_df['Age'])

#Filter rows where age is greater than 26
print(data_df[data_df['Age']>26])

#Add a new column 'Country' with the value 'India' for all rows
data_df['Country']=['India']*3
print(data_df)

data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}

df2= pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age
val=df2['Age'].mean()
df2['Age']=df2['Age'].fillna(val)
print(df2)
```

```python
import matplotlib.pyplot as plt
import numpy as np



xpoints=np.array([1,2,3,4])
ypoints=np.array([2,4,6,8])

#Plot these points without drawing a line
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(xpoints, ypoints, marker='o', linestyle='None')
plt.title('Plot without Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Plotting with star marker
plt.subplot(2, 2, 2)
plt.plot(xpoints, ypoints, marker='*', linestyle='None')
plt.title('Plot with Star Marker')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Using fmt format, add circular marker, red color, and dashed line
plt.subplot(2, 2, 3)
plt.plot(xpoints, ypoints, 'ro--', markersize=8)
plt.title('Plot with Red Circles and Dashed Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Create a scatter plot for xpoints and ypoints
plt.subplot(2, 2, 4)
plt.scatter(xpoints, ypoints, color=['blue', 'green', 'red', 'yellow'])
plt.title('Scatter Plot with Different Colors')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()
```

```python
import random
import numpy as np

#Set the seed of random to 20
random.seed(20)

arr1=np.array([1,24,31,45,73,81,94,25])

#Using the random module pick 4 different random numbers from arr1 and return their sum.
random_numbers = random.sample(list(arr1), 4)
sum_random = sum(random_numbers)

print("Randomly selected numbers:", random_numbers)
print("Sum of randomly selected numbers:", sum_random)
```
